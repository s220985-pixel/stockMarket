```python
df.dropna(axis="columns",how="all",inplace=True)
```

```python
df.shape
```

```python
numeric_cols=["Open","High","Low","Close","Volume","Daily_Change_Percent"]
```

```python
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df = df[df["Country"].str.strip() != ""]
```

```python
df.fillna(method="ffill",axis="rows",inplace=True)
```

```python
df.fillna(method="bfill",axis="rows",inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["Date"]=pd.to_datetime(df["Date"])
```

```python
df["Country"]=df["Country"].str.capitalize()
```

```python
df
```

```python
df = df[df["Date"].notna()]
df = df[df["Index_Name"].notna()]
```

```python
df.rename(columns={"Date":"Date_of_investing"},inplace=True)
```

```python
df.to_csv("cleaned_dataset1.csv",index=False)
```
